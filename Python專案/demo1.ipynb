- 這裡是一個建立 `正弦波` 的圖表

```python
import numpy as np
import matplotlib.pyplot as plt

# 建立 x 軸的數據點，從 0 到 2π，間隔為 0.1
x = np.arange(0, 2 * np.pi, 0.1)

# 使用 sin() 函數計算 y 軸的數據點
y = np.sin(x)

# 使用 plot() 函數繪製正弦波
plt.plot(x, y)

# 設定圖表的標題和軸標籤
plt.title('sin plot')
plt.xlabel('x')
plt.ylabel('sin(x)')

# 顯示圖表
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 建立 x 軸的數據點，從 0 到 2π，間隔為 0.1
x = np.arange(0, 2 * np.pi, 0.1)

# 使用 cos() 函數計算 y 軸的數據點
y = np.cos(x)

# 使用 plot() 函數繪製餘弦波
plt.plot(x, y)

# 設定圖表的標題和軸標籤
plt.title('餘弦波')
plt.xlabel('x')
plt.ylabel('cos(x)')

# 顯示圖表
plt.show()
```
